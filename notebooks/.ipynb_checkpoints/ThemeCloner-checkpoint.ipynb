# ThemeCloner
**Find small-cap (or any-universe) analogs of large-cap thematic ETFs using RP-PCA.**

Pipeline:
1. Pull ETF + target universe returns
2. Fit RP-PCA on the ETF panel to learn a theme factor space
3. Distil pure theme DNA across multiple ETFs per theme
4. Project target universe stocks into that factor space
5. Score by cosine similarity + rank candidates
6. Robustness check: compare ETF-space vs universe-native RP-PCA

Edit `config/etfs.csv` to add/remove ETFs and themes. Everything else runs automatically.

```python
# -------------------- 0. Setup --------------------
# add src/ to path so we can import modules cleanly

import sys, os
sys.path.insert(0, os.path.abspath(".."))   # if running from notebooks/
sys.path.insert(0, os.path.abspath("."))    # if running from root

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 5)
matplotlib.rcParams['axes.spines.top'] = False
matplotlib.rcParams['axes.spines.right'] = False

print('setup done')
```

## Step 1 — Pull Data

```python
# -------------------- 1. Load ETF config and pull returns --------------------

from src.data_pull import load_etf_config, pull_etf_returns, pull_universe_returns, get_russell2000_tickers

# parameters -- tweak these to change the run
START_DATE = '2018-01-01'
K          = 5      # number of RP-PCA factors
GAMMA      = 10.0   # Lettau-Pelger default -- produces good SR tradeoff
TOP_N      = 30     # candidates to show per theme

etf_config = load_etf_config('config/etfs.csv')
etf_config
```

```python
# pull ETF return series
etf_returns = pull_etf_returns(etf_config['ticker'].tolist(), start=START_DATE)
etf_returns.tail()
```

```python
# pull target universe -- Russell 2000 by default
# swap get_russell2000_tickers() for your own list if needed
universe_tickers = get_russell2000_tickers()
universe_returns = pull_universe_returns(universe_tickers, start=START_DATE)
print(f'universe: {universe_returns.shape[1]} stocks x {universe_returns.shape[0]} weeks')
```

## Step 2 — Fit RP-PCA on ETF Panel

```python
# -------------------- 2. RP-PCA on the ETF panel --------------------
# this learns the K-dimensional factor space from the ETFs
# all subsequent steps live in this same space

from src.rppca import fit_rppca, sweep_gamma, onatski_criterion

rppca_result = fit_rppca(etf_returns, K=K, gamma=GAMMA, run_oos=True)
print(f"\nloadings shape: {rppca_result['loadings'].shape}")
print(f"factors shape:  {rppca_result['factors'].shape}")
```

```python
# -------------------- gamma sweep -- confirms K and gamma choice --------------------
# this reproduces the SR-vs-gamma plot from the original replication
# we want the gamma where SR flattens out -- that's our operating point

sweep = sweep_gamma(etf_returns.fillna(0).values, K=K)

fig, ax = plt.subplots()
ax.plot(sweep['gamma'], sweep['sr'], marker='o', color='steelblue', linewidth=2)
ax.axvline(GAMMA, color='firebrick', linestyle='--', label=f'gamma={GAMMA} (chosen)')
ax.set_xlabel('gamma')
ax.set_ylabel('annualised Sharpe ratio')
ax.set_title(f'SR vs gamma  (K={K})')
ax.legend()
plt.tight_layout()
plt.savefig('outputs/fig_sr_vs_gamma.png', dpi=150)
plt.show()
```

```python
# -------------------- factor loadings heatmap --------------------
# shows which ETFs load on which factors
# useful for eyeballing whether the factors make economic sense

import seaborn as sns

fig, ax = plt.subplots(figsize=(8, 5))
sns.heatmap(rppca_result['loadings'], annot=True, fmt='.2f',
            cmap='RdBu_r', center=0, ax=ax)
ax.set_title('ETF factor loadings  (RP-PCA)')
ax.set_xlabel('factor')
ax.set_ylabel('ETF')
plt.tight_layout()
plt.savefig('outputs/fig_etf_loadings.png', dpi=150)
plt.show()
```

## Step 3 — Distil Theme DNA

```python
# -------------------- 3. Theme DNA --------------------
# for each theme, average the factor fingerprints across all ETFs in that theme
# purity score tells us how much the ETFs in a theme agree with each other
# low purity = the ETFs are quite different despite sharing a theme label

from src.theme_dna import build_theme_dna

dna_result = build_theme_dna(etf_returns, etf_config, K=K, gamma=GAMMA)

print('\npurity scores (how much ETFs within a theme agree):')
for theme, purity in dna_result['purity'].items():
    bar = '█' * int(purity * 20)
    print(f'  {theme:25s} {purity:.3f}  {bar}')
```

## Step 4 — Project Universe + Score

```python
# -------------------- 4. Project universe into theme factor space --------------------
# this is just OLS regression of each stock's returns on the K factor series
# no new RP-PCA run needed -- the factor space is already fixed from step 2

from src.projection import project_universe, score_universe, save_projections

projections = project_universe(universe_returns, etf_returns, rppca_result)
scores      = score_universe(projections, dna_result['theme_dna'])
save_projections(projections, scores)

scores.describe()
```

## Step 5 — Rank Candidates

```python
# -------------------- 5. Rank candidates per theme --------------------

from src.scoring import rank_candidates, save_outputs

ranked = rank_candidates(scores, top_n=TOP_N, min_score=0.3)
save_outputs(ranked)

# show top 10 per theme
for theme in ranked['theme'].unique():
    print(f'\n--- {theme} (top 10) ---')
    print(ranked[ranked['theme'] == theme].head(10).to_string(index=False))
```

## Step 6 — Robustness Check

```python
# -------------------- 6. Robustness: universe-native RP-PCA --------------------
# run RP-PCA on the Russell universe directly
# project ETF theme DNA into that space
# score and rank universe stocks the same way
# then compare: which stocks appear in BOTH ranked lists?

from src.rppca    import fit_rppca
from src.projection import project_universe, score_universe
from src.scoring  import rank_candidates, compare_methods

# fit RP-PCA on the universe
print('fitting RP-PCA on universe (this is the alternative method)...')
universe_rppca = fit_rppca(universe_returns, K=K, gamma=GAMMA, run_oos=False)

# project ETF returns into the universe factor space to get theme DNA in that space
etf_proj_in_universe_space = project_universe(etf_returns, universe_returns, universe_rppca)

# compute theme DNA in universe space (average across ETFs per theme)
theme_dna_universe = {}
for theme in etf_config['theme'].unique():
    tickers = etf_config[etf_config['theme'] == theme]['ticker'].tolist()
    available = [t for t in tickers if t in etf_proj_in_universe_space.index]
    if available:
        fp_matrix = etf_proj_in_universe_space.loc[available].values
        theme_dna_universe[theme] = fp_matrix.mean(axis=0)

# score universe stocks against theme DNA in universe space
universe_loadings_df = universe_rppca['loadings']  # stocks x K
scores_m2 = score_universe(universe_loadings_df, theme_dna_universe)
ranked_m2 = rank_candidates(scores_m2, top_n=TOP_N, min_score=0.3)
```

```python
# -------------------- compare both methods --------------------
# high_confidence  = agree in both  -> strong candidates
# etf_space_only   = only method 1  -> market prices them as thematic but structure differs
# universe_only    = only method 2  -> native co-movement without ETF resemblance

comparison = compare_methods(ranked, ranked_m2)
comparison[['theme', 'high_confidence', 'etf_space_only', 'universe_only', 'agreement_rate']]
```

```python
# -------------------- show high-confidence candidates --------------------
# these are the stocks both methods agree on -- strongest candidates for the paper

for _, row in comparison.iterrows():
    print(f"\n{row['theme']} -- high confidence tickers ({row['high_confidence']}):")
    print(', '.join(row['hc_tickers']) if row['hc_tickers'] else 'none')
    print(f"  ETF-space only ({row['etf_space_only']}): {', '.join(row['m1_only_tickers'][:10])}")
```
